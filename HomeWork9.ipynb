```python
import math
from sympy import*
from sympy.plotting import plot


x = Symbol('x')
fx = -5 * x ** 2 + 10 * x - 150
# fx =  -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
print("Функция")
fx
```

**Построение графика**

```python
p = plot(fx, (x, -10, 10))
```

**Нахождение вершины**

```python
diff(fx)
imp = solve(diff(fx))
x = imp[0]
y = -5 * x ** 2 + 10 * x - 150
print("Вершина")
x, y
```

**Промежутки возрастания и убывания**

```python
print("Функция растет на промежутке")
print(solve(diff(fx) > 0))
```

```python
print("Функция убывает на промежутке")
print(solve(diff(fx)<0))
```

**Промежутки, где функция < 0**

```python
print(solve(fx<0))
```

**Промежутки, где функция > 0**

```python
print(solve(fx>0))
```

**Корни уравнения**

```python
x1, x2 = solve(fx)
print('первый корень: ', x1.evalf())
print('второй корень: ', x2.evalf())
# plot(solve(fx))
```
